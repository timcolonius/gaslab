```python
from gaslab.relations import arstar, mach_from_arstar

a = arstar(2.0, 1.4)
print(a)
print(mach_from_arstar(a, 1.4, subsonic=False))
print(mach_from_arstar(a, 1.4, subsonic=True))
```

```python
from gaslab import GasState

s1 = GasState(2.0, 1.4)
s2 = s1.areachg(2.0)

print("M1 =", s1.mach)
print("M2 =", s2.mach)
print("p01 =", s1.p0, "p02 =", s2.p0)
print("T01 =", s1.t0, "T02 =", s2.t0)
```

```python
s1 = GasState(0.5, 1.4)
s2 = s1.areachg(2.0)

print("M1 =", s1.mach)
print("M2 =", s2.mach)
```

```python
s1 = GasState(3.0,1.4)
s2 = s1.areachg(0.5)     # converging duct
s3 = s2.normalshock()

for s in s3.history():
    print(s.mach, s.proc)
```
